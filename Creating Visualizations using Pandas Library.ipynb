# Creating Visualizations using Pandas Library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\ASUS\Downloads\Ice Cream Ratings.csv")
df
```

```python
df = df.set_index('Date')
df
```

```python
print(plt.style.available)
plt.style.use('fivethirtyeight')
```

```python
df.plot(kind = 'line', subplots = True);
```

```python
df.plot(kind = 'line', title = 'Ice Cream Ratings', xlabel = 'Daily Ratings', ylabel = 'Scores');
```

```python
df.plot(kind='bar');
```

```python
df.plot(kind='bar', stacked = True);
```

```python
df.plot.barh(stacked = True);
```

```python
df['Flavor Rating'].plot.barh(stacked = True);
```

```python
df['Flavor Rating'].plot(kind = 'bar');
```

```python
df.plot.scatter(x= 'Texture Rating', y = 'Overall Rating', s = 200, c = 'r');
```

```python
df.plot.hist(bins=20);
```

```python
df.boxplot();
```

```python
df.plot.area(figsize = (10,5));
```

```python
df.plot.pie(y = 'Flavor Rating', figsize = (10,6));
```
